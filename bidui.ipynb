```python
def my_map(arg):
    if arg is np.NaN:
        return ''
    try:
        resu = res1.loc[arg,'PGD结果']
        return resu
    except:
        return '未匹配到'
```

```python
def round7(arg):
    if isinstance(arg,float) or isinstance(arg,int):
        return round(arg,1)
    return arg

def round8(arg):
    if isinstance(arg,float) or isinstance(arg,int):
        return round(arg,6)
    return arg

def round9(arg):
    #round()只能转数字，故作判断
    if isinstance(arg,float) or isinstance(arg,int):
#         ValueError: cannot convert float NaN to integer，//round（）不能把np.NaN转成整数型的，故作判断
        if math.isnan(arg):
            return np.NAN
        return round(arg)
    return arg

def round10(arg):
    if isinstance(arg,float) or isinstance(arg,int):
        return round(arg,1)
    return arg

def round11(arg):
    if isinstance(arg,float) or isinstance(arg,int):
        return round(arg,1)
    return arg

def round12(arg):
    if isinstance(arg,float) or isinstance(arg,int):
        return round(arg,1)
    return arg

def round13(arg):
    if isinstance(arg,float) or isinstance(arg,int):
        return round(arg,1)
    return arg

def round15(arg):
    if isinstance(arg,float) or isinstance(arg,int):
        if math.isnan(arg):
            return np.NAN
        return round(arg)
    return arg

def round16(arg):
    if isinstance(arg,float) or isinstance(arg,int):
        if math.isnan(arg):
            return np.NAN
        return round(arg)
    return arg
```

```python
tasklist2[7] = tasklist2[7].map(round7)
tasklist2[8] = tasklist2[8].map(round8)
tasklist2[9] = tasklist2[9].map(round9)
tasklist2[10] = tasklist2[10].map(round10)
tasklist2[11] = tasklist2[11].map(round11)
tasklist2[12] = tasklist2[12].map(round12)
tasklist2[13] = tasklist2[13].map(round13)
tasklist2[15] = tasklist2[15].map(round15)
tasklist2[16] = tasklist2[16].map(round16)
```
